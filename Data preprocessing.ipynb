### Imputer object using the mean strategy and missing values type for imputation.

#### Making copy of the data to avoid changing original data (during imputing)

```python
new_df=df.copy()
```

#### Making new columns indicating what will be imputed

```python
cols_with_missing = (col for col in new_df.columns
                    if new_df[col].isnull().any())
for col in cols_with_missing:
    new_df[col] = new_df[col].isnull()
```

#### Imputation

```python
my_imputer = SimpleImputer()
new_df = pd.DataFrame(my_imputer.fit_transform(new_df))
new_df.columns = df.columns
```
